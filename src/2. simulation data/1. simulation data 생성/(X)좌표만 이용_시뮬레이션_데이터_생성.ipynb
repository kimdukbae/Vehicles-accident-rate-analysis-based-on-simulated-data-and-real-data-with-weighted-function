```python
from random import *
import pandas as pd
```

```python
def createStartEnd(df):
    for i in range(50000):
        df.loc[i] = [(randint(100,150), randint(0,180)), (randint(0,150), randint(180,270))]
```

```python
# 데이터 프레임 생성
df = pd.DataFrame(data={'start':[0,0], 'end':[0,0]})
df = df[df.start != 0]

createStartEnd(df)
```

```python
print(df)
```

```python
def computePorF(df):
    PorF = []
    for i in range(len(df)):
        x2 = df['end'][i][0]
        y2 = df['end'][i][1]
        
        if ((x2 > 0 and x2 < 100 and y2 > 200 and y2 < 250) or (x2 > 100 and x2 < 150 and y2 > 180 and y2 < 250)):
            PorF.append(0)
        else:
            PorF.append(1)
            
    success = pd.Series(PorF)
    return success
```

```python
success = computePorF(df)
```

```python
success
```

```python
df['PorF'] = success
```
