```python
import random

def createNewData(df, end):
    temp = df

    nd = pd.DataFrame(index=range(0,0), columns=['POSTED_SPEED_LIMIT', 'WEATHER_CONDITION', 'ROADWAY_SURFACE_COND', 'ROAD_DEFECT', 'CRASH_HOUR', 'speed', 'angle'])

    x1 = list(temp['POSTED_SPEED_LIMIT'])
    x2 = list(temp['WEATHER_CONDITION'])
    x3 = list(temp['ROADWAY_SURFACE_COND'])

    x4 = list(temp['ROAD_DEFECT'])
    del x4[2:]

    x5 = list(temp['CRASH_HOUR'])
    del x5[4]

    x6 = list(temp['speed'])

    x7 = list(temp['angle'])
    del x7[4]

    for i in range(end):
        row = []
        row.append(random.choice(x1))
        row.append(random.choice(x2))
        row.append(random.choice(x3))
        row.append(random.choice(x4))
        row.append(random.choice(x5))
        row.append(random.choice(x6))
        row.append(random.choice(x7))

        nd.loc[i] = row

    return nd
```

```python
new_data = createNewData(table, 1000)
```

```python
def computeY(df):
    fx1 = []
    fx2 = []
    
    for i in range(len(df)):
        res1 = 0.75*df['POSTED_SPEED_LIMIT'][i] + 0.1375*df['WEATHER_CONDITION'][i] + 0.1425*df['ROADWAY_SURFACE_COND'][i] + 0.1375*df['ROAD_DEFECT'][i] + 0.1775*df['CRASH_HOUR'][i] + 0.0225*df['speed'][i] + 0.1075*df['angle'][i]
        res2 = 0.0546*df['POSTED_SPEED_LIMIT'][i] + 0.1093*df['WEATHER_CONDITION'][i] + 0.1093*df['ROADWAY_SURFACE_COND'][i] + 0.1694*df['ROAD_DEFECT'][i] + 0.1913*df['CRASH_HOUR'][i] + 0.2404*df['speed'][i] + 0.1257*df['angle'][i]
        
        fx1.append(res1)
        fx2.append(res2)
        
    fx1 = pd.Series(fx1)
    fx2 = pd.Series(fx2)
    
    df['fx1'] = fx1
    df['fx2'] = fx2
```

```python
computeY(model)
```
