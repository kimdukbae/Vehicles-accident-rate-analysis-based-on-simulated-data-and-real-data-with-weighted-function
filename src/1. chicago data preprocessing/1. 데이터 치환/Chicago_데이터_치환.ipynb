# 2. TRAFFIC_CONTROL_DEVICE

```python
# DataFrame 복사
transform = chicago

def change2(x):
    if x=="NO CONTROLS": return 1
    elif x=="TRAFFIC SIGNAL": return 2
    elif x=="STOP SIGN/FLASHER": return 3
    elif x=="OTHER": return 4
    elif x=="LANE USE MARKING": return 5
    elif x=="YIELD": return 6
    elif x=="OTHER REG. SIGN": return 7
    elif x=="OTHER WARNING SIGN": return 8
    elif x=="RAILROAD CROSSING GATE": return 9
    elif x=="POLICE/FLAGMAN": return 10
    elif x=="SCHOOL ZONE": return 11
    elif x=="PEDESTRIAN CROSSING SIGN": return 12
    elif x=="DELINEATORS": return 13
    elif x=="OTHER RAILROAD CROSSING": return 14
    elif x=="FLASHING CONTROL SIGNAL": return 15
    elif x=="RR CROSSING SIGN": return 16
    elif x=="NO PASSING": return 17
    elif x=="BICYCLE CROSSING SIGN": return 18
    elif x=="UNKNOWN": return 0
    
transform["TRAFFIC_CONTROL_DEVICE"] = transform["TRAFFIC_CONTROL_DEVICE"].apply(change2)
```

```python
transform
```

```python
transform["TRAFFIC_CONTROL_DEVICE"].value_counts(normalize=True) * 100
```
