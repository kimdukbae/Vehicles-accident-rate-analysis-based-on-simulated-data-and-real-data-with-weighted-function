# 1. POSTED_SPEED_LIMIT

```python
def change1(x):
    if x > 0 and x <= 10: return 1
    elif x > 10 and x<= 20: return 2
    elif x > 20 and x<= 30: return 3
    elif x > 30 and x<= 40: return 4
    else: return 5
    
transform["POSTED_SPEED_LIMIT"] = transform["POSTED_SPEED_LIMIT"].apply(change1)
```

```python
transform["POSTED_SPEED_LIMIT"].value_counts(normalize=True) * 100
```

```python
def change1(x):
    if x==1: return 0.03
    elif x==2: return 0.07
    elif x==3: return 0.8
    elif x==4: return 0.08
    else: return 0.02
    
transform["POSTED_SPEED_LIMIT"] = transform["POSTED_SPEED_LIMIT"].apply(change1)
```
